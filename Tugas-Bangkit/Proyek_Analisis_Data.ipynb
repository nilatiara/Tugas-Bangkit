### Pertanyaan 2: Bagaimana pengaruh suhu (temp) terhadap penyewaan sepeda di berbagai musim (season)?

```python
sns.scatterplot(x='temp', y='cnt', hue='season', data=df_day)
plt.title('Pengaruh Suhu terhadap Penyewaan Sepeda di Berbagai Musim')
plt.xlabel('Suhu (Normalisasi)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.legend(title='Season', labels=['Spring', 'Summer', 'Fall', 'Winter'])
plt.show()
```

**Insight:**

- Setiap titik pada plot akan diwarnai berdasarkan musim:
1 = Spring (Musim Semi)
2 = Summer (Musim Panas)
3 = Fall (Musim Gugur)
4 = Winter (Musim Dingin)


## Analisis Lanjutan (Opsional)

```python
import datetime

df_day['dteday'] = pd.to_datetime(df_day['dteday'])

last_date = df_day['dteday'].max()

df_day['Recency'] = (last_date - df_day['dteday']).dt.days

df_day['R_Quartile'] = pd.qcut(df_day['Recency'], 4, labels=False)
df_day['F_Quartile'] = pd.qcut(df_day['cnt'], 4, labels=False)
df_day['M_Quartile'] = pd.qcut(df_day['cnt'], 4, labels=False)  # Menggunakan cnt sebagai proxy Monetary

df_day['RFM_Segment'] = df_day['R_Quartile'].astype(str) + df_day['F_Quartile'].astype(str) + df_day['M_Quartile'].astype(str)

print(df_day.groupby('RFM_Segment')['cnt'].mean())
```

```python
import random
from sklearn.preprocessing import StandardScaler

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1-x2)**2))

def kmeans(data, k, max_iters=100):
    centroids = data[random.sample(range(len(data)), k)]

    for _ in range(max_iters):
        clusters = [np.argmin([euclidean_distance(x, centroid) for centroid in centroids]) for x in data]

        new_centroids = [data[np.where(np.array(clusters) == i)].mean(axis=0) for i in range(k)]

        if np.allclose(centroids, new_centroids):
            break

        centroids = new_centroids

    return centroids, clusters

X = df_day[['temp', 'hum', 'windspeed', 'cnt']].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

centroids, clusters = kmeans(X_scaled, k=3)

df_day['cluster'] = clusters

print(df_day.groupby('cluster')[['temp', 'hum', 'windspeed', 'cnt']].mean())
```
